## Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings("ignore")
```

## Data

```python
path = '../Datasets/creditcard.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.drop(['Time'],axis=1, inplace=True)
```

```python
scaler = StandardScaler()
df['Amount'] = scaler.fit_transform(pd.DataFrame(df['Amount']))
df.head()
```

```python
df['Class'].value_counts()
```

```python
sns.countplot(data=df, x='Class')

plt.title("Count of Classes")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

## X and y variables

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

## Logistic Regression

```python
lr = LogisticRegression(class_weight="balanced", solver="lbfgs", max_iter=1000, random_state=42)
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)

print(classification_report(y_test, y_pred))

lr_cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(lr_cm, annot=True, fmt="d", cmap="Blues", xticklabels=[0,1], yticklabels=[0,1])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix - Logistic Regression")
plt.show()
```

## Decision Tree Classifier

```python
dtc = DecisionTreeClassifier(class_weight="balanced", random_state=42)
dtc.fit(X_train, y_train)
```

```python
y_pred = dtc.predict(X_test)

print(classification_report(y_test, y_pred))

dtc_cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(dtc_cm, annot=True, fmt="d", cmap="Blues", xticklabels=[0,1], yticklabels=[0,1])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix - Decision Tree Classifier")
plt.show()
```

## Random Forest Classifier

```python
rfc = RandomForestClassifier(class_weight="balanced", random_state=42, n_estimators=100)
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)

print(classification_report(y_test, y_pred))

rfc_cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(rfc_cm, annot=True, fmt="d", cmap="Blues", xticklabels=[0,1], yticklabels=[0,1])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix - Random Forest Classifier")
plt.show()
```

## Support Vector Classifier

```python
svc = SVC(class_weight="balanced", random_state=42)
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)

print(classification_report(y_test, y_pred))

svc_cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(svc_cm, annot=True, fmt="d", cmap="Blues", xticklabels=[0,1], yticklabels=[0,1])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix - Support Vector Classifier")
plt.show()
```

## With GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
models = {
    "LogisticRegression": (LogisticRegression(max_iter=1000), {
        "C": [0.01, 0.1, 1, 10],
        "solver": ["liblinear", "lbfgs"]
    }),
    "DecisionTree": (DecisionTreeClassifier(), {
        "criterion": ["gini", "entropy"],
        "max_depth": [None, 5, 10, 20]
    }),
    "RandomForest": (RandomForestClassifier(), {
        "n_estimators": [50, 100, 200],
        "max_depth": [None, 5, 10, 20]
    }),
    "SVC": (SVC(), {
        "C": [0.1, 1, 10],
        "kernel": ["linear", "rbf"],
        "gamma": ["scale", "auto"]
    })
}

for name, (model, params) in models.items():
    print(f"===== {name} =====")
    grid = GridSearchCV(model, params, cv=5, scoring="accuracy", n_jobs=-1)
    grid.fit(X_train, y_train)

    best_model = grid.best_estimator_
    y_pred = best_model.predict(X_test)

    print("Best Parameters:", grid.best_params_)
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))

    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"{name} Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
```
